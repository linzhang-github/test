```python
import pandas as pd
import keras
import tensorflow as tf
import string
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

```python
stop_words = set(stopwords.words("english"))
ps = PorterStemmer()
```

```python
reviews = pd.read_csv('df_restaunt_with_topics.csv')
```

```python
reviews = reviews[['review_rating','text','review_id']].set_index('review_id',drop=True)
reviews
```

```python
reviews['text'] = list(map(lambda x: x.translate(str.maketrans('','',string.punctuation)),reviews['text']))
reviews['text'] = list(map(lambda x: x.translate(str.maketrans('','',string.digits)),reviews['text']))
reviews['text'] = list(map(lambda x: x.lstrip().rstrip().lower(),reviews['text']))
```

```python
reviews = reviews.sample(frac=1) #shuffles rows

reviews_train = reviews.head(len(reviews)-99903)
reviews_test = reviews.tail(99903)

y_train = reviews_train['review_rating']
y_test = reviews_test['review_rating']
```

```python
vocabulary_size = 5000

vectorizer = TfidfVectorizer(lowercase = True, decode_error='ignore', sublinear_tf=True, preprocessor = ps.stem,stop_words = stop_words,max_features=vocabulary_size,ngram_range=(1,1))
```

```python
reviews_td_matrix = vectorizer.fit_transform(reviews['text'])
X_train = vectorizer.fit_transform(reviews_train['text'])
X_test = vectorizer.fit_transform(reviews_test['text'])
```

```python
from keras.preprocessing import sequence
max_words = 500
X_train = sequence.pad_sequences(X_train, maxlen=max_words)
X_test = sequence.pad_sequences(X_test, maxlen=max_words)
```

```python
from keras import Sequential
from keras.layers import Embedding, LSTM, Dense, Dropout
embedding_size=32
model=Sequential()
model.add(Embedding(vocabulary_size, embedding_size, input_length=max_words))
model.add(LSTM(100))
model.add(Dense(1, activation='sigmoid'))
print(model.summary())
```

```python
model.compile(loss='binary_crossentropy', 
             optimizer='adam', 
             metrics=['accuracy'])
```

```python
batch_size = 64
num_epochs = 5
X_valid, y_valid = X_train[:batch_size], y_train[:batch_size]
X_train2, y_train2 = X_train[batch_size:], y_train[batch_size:]
model.fit(X_train2, y_train2, validation_data=(X_valid, y_valid), batch_size=batch_size, epochs=num_epochs)
```

```python
scores = model.evaluate(X_test, y_test, verbose=0)
print('Test accuracy:', scores[1])
```
